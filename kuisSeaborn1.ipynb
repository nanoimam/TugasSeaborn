buatlah data diagram seprti ini

```python
import seaborn as sns
data = sns.load_dataset("tips")
data
```

```python
sns.catplot(x="day", y="total_bill", data=data, hue="smoker", kind="violin", inner="stick",palette="Set1")
```

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(x="color", y="price", data=diamonds, kind="bar")
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(y="deck",hue="sex", data=titanic, kind="count", palette="Set1")
```
